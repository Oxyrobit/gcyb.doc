Exercice advent of code 01 de 2015

### Objectif :

On vous fournit un string contenant des instructions sur le format suivant : "(())(()". L'idée est qu'on et au rez de chaussé et on va monter et descendre en suivant les instructions et on doit donner le niveau final ou on arrive à la fin.

Il y a deux types d'instructions :
- ( signifie je monte d'un niveau
- ) signifie je descend d'un niveau

### Consigne :

1. Parcourir les caractères de la chaine d'instruction
2. selon le caractère, monter ou descendre d'un niveau
3. Un fois toutes les instructions suivies, afficher le niveau.

````python
 instruction = "((((()(()(((((((()))(((()((((()())(())()(((()((((((()((()(()(((()(()((())))()((()()())))))))))()((((((())("

floor = 0
for order in instruction:
    if order == "(":
        floor += 1
    else:
        floor -= 1
print(f" L'index : {floor}")
# Commence par les comentaires pour décomposer le problème
````

1. Trouver le dernier nombre de la chaine (2 pour "1abc2", 8 pour "pqr3stu8vwx" et 7 pour "treb7uchet")
1. Assembler les deux selon le schéma suivant : (12 pour "1abc2", 38 pour "pqr3stu8vwx" et 77 pour "treb7uchet")
1. Calcul la somme des nombres ( 12 + 38 + 15 + 77 = 142)
1. Afficher la somme

Résultat attendu :

    142

Astuce


````python
test = "R0m@1n"
print(f"Test: {test}")
for char in test:
    if char.isdigit():
        print(f"Le caractère {char} est un digit")
    else:
        print(f"Le caractère {char} n'est pas un digit")
````

````python
dataChallenge = ["1abc2","pqr3stu8vwx","a1b2c3d4e5f","treb7uchet", "azerthjfgjbhkd"]

def GetDigit(string:str):
    digit = ''
    for char in string:
        if char.isdigit():
            digit += char
    return digit

total_count = 0
for data in dataChallenge:
    number_string = GetDigit(data)
    if number_string == '': continue
    first_last_number = number_string[0] + number_string[len(number_string) - 1]
    total_count += int(first_last_number)

#print(total_count)
````

````python
# Utilisation des données test
dataChallenge = []
dataChallenge.append("3 4")
dataChallenge.append("4 3")
dataChallenge.append("2 5")
dataChallenge.append("1 3")
dataChallenge.append("3 9")
dataChallenge.append("3 3")

liste = [[],[]]
for data in dataChallenge:
    liste[0].append(data[0])
    liste[1].append(data[2])

liste[0].sort()
liste[1].sort()

total_count = 0
for i in range(len(liste[0])):
    diff = abs(int(liste[0][i]) - int(liste[1][i]))
    total_count += diff

print(total_count)
````
